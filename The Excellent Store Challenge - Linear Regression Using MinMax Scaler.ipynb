# Fill up Missing Values

Categrorical Values

```python
# Fill up missing categorical values ('TOP_PACK' and 'REGION' ) with N/A
for i in ['Store_Size']:
    df_train[i] = df_train[i].fillna(value = 'N/A')
    df_test[i] = df_test[i].fillna(value = 'N/A')
```

```python
# Fill up missing numerical values with the mean
for i in ['Item_Weight']:
    df_train[i] = df_train[i].fillna(value = 0)
    df_test[i] = df_test[i].fillna(value = 0)
```

This will be done on "Item_ID", "Store_ID", "Item_Sugar_Content", "Item_Type", "Store_Size", "Store_Location_Type", 
"Store_Type".

"Item_Store_ID" will be dropped because all entries are unique, so no effect on the expected result.

```python
from sklearn.preprocessing import LabelEncoder
```

```python
#For Train Dataset
le = LabelEncoder()
df_train["Item_ID"] = le.fit_transform(df_train["Item_ID"].astype(str))
df_train["Store_ID"] = le.fit_transform(df_train["Store_ID"].astype(str))
df_train["Item_Sugar_Content"] = le.fit_transform(df_train["Item_Sugar_Content"].astype(str))
df_train["Item_Type"] = le.fit_transform(df_train["Item_Type"].astype(str))
df_train["Store_Size"] = le.fit_transform(df_train["Store_Size"].astype(str))
df_train["Store_Location_Type"] = le.fit_transform(df_train["Store_Location_Type"].astype(str))
df_train["Store_Type"] = le.fit_transform(df_train["Store_Type"].astype(str))
df_train
```

```python
#For Test Dataset
le = LabelEncoder()
df_test["Item_ID"] = le.fit_transform(df_test["Item_ID"].astype(str))
df_test["Store_ID"] = le.fit_transform(df_test["Store_ID"].astype(str))
df_test["Item_Sugar_Content"] = le.fit_transform(df_test["Item_Sugar_Content"].astype(str))
df_test["Item_Type"] = le.fit_transform(df_test["Item_Type"].astype(str))
df_test["Store_Size"] = le.fit_transform(df_test["Store_Size"].astype(str))
df_test["Store_Location_Type"] = le.fit_transform(df_test["Store_Location_Type"].astype(str))
df_test["Store_Type"] = le.fit_transform(df_test["Store_Type"].astype(str))
df_test
```

# Normalising the Dataset To A Common Scale Using The Min Max Scaler



Drop Item "Item_Store_ID"

```python
df_train = df_train.drop(["Item_Store_ID"], axis =1)
```

```python
scaler = MinMaxScaler()
normalised_df_train = pd.DataFrame(scaler.fit_transform(df_train), columns=df_train.columns)
```

# Build Model


Declare Label and Features

```python
# Recollect "Item_Store_ID" is to be dropped
X_train = normalised_df_train.drop(["Item_Store_Returns"], axis =1)
y_train = normalised_df_train["Item_Store_Returns"]

X_test = df_test.drop(["Item_Store_ID"], axis = 1)
```

Split Training and Test Data

```python
x_train,x_validation,y_train,y_validation = train_test_split(X_train,y_train,test_size=0.30,random_state=7)
```

Declare Model Type

```python
linear_model = LinearRegression()
```

Fit the model to the training dataset

```python
linear_model.fit(x_train,y_train)
```

Model Accuracy

```python
linear_model.score(x_validation,y_validation)
```

```python
prediction = linear_model.predict(x_validation)
prediction
```

# Measuring Performance

Root Mean Squared Error

```python
rmse = np.sqrt(mean_squared_error(y_validation, prediction))
round(rmse, 3)
```

Making Predictions on Test Dataset

```python
Test_prediction = linear_model.predict(X_test)[2] #predicting probability.  showing the user_id column and the prediction alone
```

Save prediction file for submission

```python
pd.DataFrame({"Item_Store_ID": df_test["Item_Store_ID"], "Item_Store_Returns": Test_prediction}).to_csv("First_Submission.csv", index = False)
```
